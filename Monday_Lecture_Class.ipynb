### If Statements

```python
# Available operators Greater than (>), less than (<), (==)
# not equal to (!=) or append the word not
#greater or equal (>=) and less than or equal (<=)

num1 = 5
num2 = 10
if num1 == num2:
    print("equal")
else:
    print("Not equal")
    
if num1 >= num2:
    print("num1 is greater")
else:
    print("num2 is greater")
    
#and/or

if num1 >= 5 and num2 < 15:
    print("5 is less than or equal to num1 and 15 is greater than num2")
else:
    print("opposite")
    
if num1 >= 10 and num2 < 15:
    print("10 is less than or equal to num1 and 15 is greater than num2")
else:
    print("opposite")
    
# if True or True
if num1 >= 5 or num2 < 15:
    print("5 is less than or equal to num1 and 15 is greater than num2")
else:
    print("opposite")
#if False or True
if num1 >= 10 or num2 < 15:
    print("10 is less than or equal to num1 and 15 is greater than num2")
else:
    print("opposite")
    
if not 5 == 4:
    print("not equal")
if 5 != 4:
    print("not equal")
```

## Exercise #1 <br>
<p>Cube Number Test... Print out all cubed numbers up to the total value 1000, so if the cubed number is over 1000 break the loop.</p>

```python
one_thousand = range(1,1000)

for num in one_thousand:
    cubed = num ** 3
    
    if cubed > 1000:
        break
    print(cubed)
```

## Exercise #2 <br>
<p>Get first prime numbers up to 100</p>

```python
lower = 1
upper = 100

for num in range(lower, upper + 1):
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num)
```
